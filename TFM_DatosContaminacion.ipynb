```python
import os
```

```python
import pandas as pd
```

```python
path = "C:\\Users\\Aeroengy\\Downloads"
```

```python
carpetas = os.listdir(path)
carpetas.remove('desktop.ini')
carpetas
```

```python
lista_csv = []
for año in carpetas:
    files = os.listdir(path + '\\' + año)
    lista_csv = lista_csv + files
lista_csv
```

```python
estacion = []
año = []

for medicion in lista_csv:
    estacion.append(medicion[10:13])
    año.append(medicion[13:17])   
```

```python
d = {'estacion':estacion, 'año':año}
df = pd.DataFrame(d)
df
```

```python
df.groupby('estacion').count().sort_values('estacion')
```
