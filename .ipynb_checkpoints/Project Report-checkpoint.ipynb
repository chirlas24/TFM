## 5. Run the model

```python
data_path = '/home/gonzalo/Data/TFM/DATA/'
```

### 5.1. Load the data

```python
#Libraries
import project as p
import pandas as pd
from datetime import datetime
import os
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.model_selection import GridSearchCV
import altair as alt
alt.renderers.enable('notebook')
import json
import geopandas as gpd
import pickle
```

```python
#Load data
df_dock_bikes = p.make_all_dataset(data_path, 'dock_bikes',by_postal_code=True, verbose=False )
df_total_bases = p.total_bases_dataset(data_path)
df_occupation_rate = p.make_df_occupation_rate(df_dock_bikes, df_total_bases)
```

```python
#Apply feature engineering
df_weather = pd.read_json(os.path.join(data_path, 'METEO.txt'))
df = p.feature_engineering(df_occupation_rate, df_weather)
```

### 5.2. Model

```python
#Load the model
model = pickle.load(open('model', 'rb'))
```

```python
#Divide the data
historical = df.iloc[:-1]
to_predict = df.iloc[-1:]
```

```python
#Predict
X, y = p.get_X_y(to_predict)
y_predicted = model.predict(X)
```

```python
#Columns for labels and features
columns = df.columns
X_columns = [x for x in columns if '28' in x and 'lag' in x]
X_columns = X_columns + ['prec', 'presMax', 'tmed', 'velmedia', 'day_month', 'back_days', 'sin_weekday', 'cos_weekday']
y_columns = [y for y in columns if '28' in y and 'lag' not in y]
```

```python
#Remake the dataframe with the predicted values
to_predict[y_columns] = y_predicted
df = pd.concat([historical, to_predict])
```

### 5.3. Plot

```python
#Make the plot dataframe
columns = [x for x in df.columns if '28' in x and not 'lag' in x]
df = df[columns]
df.columns = [x.split('_')[0] for x in df.columns]
df.index = df_occupation_rate.index
df = df.T
df = p.pivot_data_frame(df)
df_coordinates_postal_codes = pd.read_csv(os.path.join(data_path, 'CoordinatesPostalCodes.csv'), sep=';', dtype={'Codigo Postal' : str})
df_coordinates_postal_codes.set_index('Codigo Postal', inplace=True)
df = df.merge(df_coordinates_postal_codes, how='left', left_on='Postal Code', right_index=True)
```

```python
#Choose prediction
df_plot_prediction = df[df['Date'] == df['Date'].max()]
```

```python
#Add the number of stations by postal code
stations_postal_code = p.load_dict_from_file(data_path)

df_stations_postal_code = pd.DataFrame()
list_aux = []

for postal_code in stations_postal_code:
    count_stations = len(stations_postal_code[postal_code])
    list_aux.append(count_stations)
    df_stations_postal_code[postal_code] = list_aux
    list_aux = []
    
df_stations_postal_code = df_stations_postal_code.T
df_stations_postal_code.columns = ['n_stations']
df = df.merge(df_stations_postal_code, how='left', left_on='Postal Code', right_index=True)
df_plot_prediction = df_plot_prediction.merge(df_stations_postal_code, how='left', left_on='Postal Code', right_index=True)
```

```python
p.plot(df, df_plot_prediction, data_path)
```
